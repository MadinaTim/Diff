```python
df = df.query("Country == 'Colombia' & Year >= 1991 & Year <= 2010 & Attribute == 'TFP_Index'")
```

```python
df['Value'] = df['Value'].str.replace(',', '.').astype(float)
```

```python
df['growth'] = df.Value.pct_change()*100
```
